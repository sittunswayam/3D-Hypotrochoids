```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from mpl_toolkits.mplot3d import Axes3D
import imageio
%matplotlib notebook
```

```python
R = 4
r = 1.8
d = 1.2
n = 5000
a = 1
```

```python
theta = np.linspace(0, 16*np.pi, n+1)
```

```python
x = ((R-r)*np.cos(theta) + d*np.cos(((R-r)/r)*theta))
y = ((R-r)*np.sin(theta) - d*np.sin(((R-r)/r)*theta))*(np.cos(a*theta))
z = ((R-r)*np.sin(theta) - d*np.sin(((R-r)/r)*theta))*(np.sin(a*theta))
```

```python
import os
```

```python
fig = plt.figure(figsize=(6,6))
images = []

for i in range(181):
    ax = plt.axes(projection='3d')
    ax.set_xlim(-3,3)
    ax.set_ylim(-3,3)
    ax.set_zlim(-3,3)
    ax.plot3D(x,y,z,linewidth=0.5,c='c')
    ax.plot3D(x,y,-z,linewidth=0.5,c='c')
    ax.plot3D(y,x,z,linewidth=0.5,c='c')
    ax.plot3D(y,x,-z,linewidth=0.5,c='c')
    plt.axis('off')
    ax.set_facecolor('k')
    ax.view_init(elev=0, azim=i)
    plt.legend([f'azimuth = {i}'], loc=3)
    if (i >= 0 and i <= 9):
        plt.savefig("Frames\movie00%d.png" % i)
    elif (i >= 10 and i <= 99):   
        plt.savefig("Frames\movie0%d.png" % i)
    elif (i >= 100 and i <= 180):   
        plt.savefig("Frames\movie%d.png" % i)   
```

```python
for file in os.listdir('Frames'):
    images.append(imageio.imread('Frames\{}'.format(file)))
    os.remove('Frames\{}'.format(file))
imageio.mimsave('movie.gif', images, fps=45)
```
